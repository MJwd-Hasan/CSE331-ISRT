## **Importing Essentials**

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings("ignore")
def solution(n):
    print('='*15)
    print('|||', f'Solution {n}')
    print('='*15)
def lb():
    print('\n')
```

## **Solutions**

```python
solution(1)
try:
    df = pd.read_csv('customer_data.csv')
    print("Data loaded successfully.")
except FileNotFoundError:
    print("Error: 'CustomerData.csv' file not found in the current directory.")
    df = pd.DataFrame()
lb()

solution(2)
print('The first 5 ros of the Customer Data are as follows:\n')
print(df.head(5))
df['full_name'] = df['prefix'] + ' ' + df['first'] + ' ' + df['last']
lb()
solution(3)
print(f'Customer Data has {df.shape[0]} rows and {df.shape[1]} columns.\n')
lb()

solution(4)
print('The maximum, minimum and average age of the customers are as follows:')
print(df['age'].describe()[['max', 'min', 'mean']])
lb()

```

```python
solution(5)
print(f'There are total {df['phone'].nunique()} phone numbers. So only {df.shape[0] - df['phone'].nunique()} duplicate phone numbers exist in the data.')
customer_with_duplicate_phones = df[df.duplicated(['phone'], keep=False)].sort_values(by='phone')['full_name'].tolist()
print('List of customers with duplicate phone numbers:')
for name in customer_with_duplicate_phones:
    print(name)
lb()

solution(6)
print(f'There are total {df[df['profession'] == 'Structural Engineer'].value_counts().sum()} Structural Engineers in the data.')
lb()

solution(7)
print(f'There are total {df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Male')].value_counts().sum()} male customers who are Structural Engineers.')
lb()

solution(8)
print(f'There are total {df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Female')].value_counts().sum()} female customers who are Structural Engineers.')
lb()


solution(9)
print(f'There are total {df[(df['profession'] == 'Structural Engineer') & (df['gender'] == 'Female') & (df['province'] == 'AB')].value_counts().sum()} female customers who are Structural Engineers from the province of Alberta(AB).')
lb()

solution(10)
print('The maximum, minimum and average spending of the customers are as follows:')
print(df['price(CAD)'].describe()[['max', 'min', 'mean']])
lb()
```

```python
solution(11)
print(f'There are total {df[df['price(CAD)'] == 0].value_counts().sum()} customers who have spent 0 CAD.')
print('List of customers who have spent 0 CAD:')
no_spending = df[df['price(CAD)'] == 0][['full_name', 'phone']].reset_index(drop = True)
print(no_spending)
lb()

solution(12)
total_spending = df.groupby('full_name').agg({'price(CAD)': 'sum', 'phone': 'first'}).sort_values(by = 'price(CAD)', ascending = False).reset_index()
above_100 = total_spending[total_spending['price(CAD)'] > 100]
print(above_100)
lb()
```

```python
solution(13)
df_cc = df[df['cc_no'] == 5020000000000230]

result = df_cc[df_cc.duplicated('email', keep=False)][['full_name', 'email', 'cc_no']]

if not result.empty:
    print("List of customers having the same email and credit card number 5020000000000230:")
    print(result.reset_index(drop=True))
else:
    print("There are no customers with the same email and credit card number 5020000000000230.")
lb()

solution(14)
# fetching the year of expiring
df['cc_exp'] = pd.to_datetime(df['cc_exp'])
df['cc_exp_year'] = df['cc_exp'].dt.year
print(f'There are total {df[df['cc_exp_year'] == 2019 ].count().sum()} which were expired in 2019')
lb()


solution(15)
print(f'{df[df['cc_type'] == 'Visa'].count().sum()} people use Visa as their Credit Card Provider')
```

```python
solution(16)
df['total_spending'] = df.groupby('full_name')['price(CAD)'].transform('sum')
query15 = df[(df['cc_type']=='Visa') & (df['total_spending']==100)]
print(f'{query15.count().sum()} number of customers spent 100 CAD using Visa')
lb()

solution(17)
professions_count_sorted = df['profession'].value_counts(sort = True).reset_index()
print(f'Two most common professions are:')
for rank in range(2):
    print(f'{rank+1}. {professions_count_sorted.iloc[rank, 0]}')
lb()

solution(18)
df['provider'] = df['email'].str.split('@').str[1]
provider_counts = df['provider'].value_counts(sort = True).reset_index()
print(f'Five most common professions are:')
for rank in range(5):
    print(f'{rank+1}. {provider_counts.iloc[rank, 0]}')
lb()

solution(19)
query19 = df[df['provider']=='am.edu'].count().sum()
if query19  > 0:
    print(f'Yes, there are {query19} customers who use am.edu as their email provider.')
else:
    print(f'No, there are no customers who use am.edu as their email provider.')
lb()

solution(20)
query20 = df['weekday'].value_counts(sort = True).reset_index()
print(f'The store gets more customers on {query20.iloc[0,0]}.')
```
